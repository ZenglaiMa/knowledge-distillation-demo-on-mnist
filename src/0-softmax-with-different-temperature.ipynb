探究 softmax 在不同 temperature 下的效果。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x, temperature):
    return np.exp(x / temperature) / np.sum(np.exp(x / temperature))
```

```python
logits = np.array([-5, 2, 7, 9])
```

```python
softmax_1 = softmax(logits, temperature=1)
softmax_3 = softmax(logits, temperature=3)
softmax_5 = softmax(logits, temperature=5)
softmax_10 = softmax(logits, temperature=10)
softmax_100 = softmax(logits, temperature=100)

plt.plot(softmax_1, label='$temperature=1$')
plt.plot(softmax_3, label='$temperature=3$')
plt.plot(softmax_5, label='$temperature=5$')
plt.plot(softmax_10, label='$temperature=10$')
plt.plot(softmax_100, label='$temperature=100$')

plt.legend()
plt.show()
```
